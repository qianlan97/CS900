```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data from the table
layers = ['Layer 4', 'Layer 5', 'Layer 6', 'Layer 29']
layer_types = ['Conv2d', 'BatchNorm2d', 'ReLU', 'BatchNorm2d']
sdc_prob = [0.0814, 0.9999, 0.0002, 0.7175]
data_size = [3211264, 3211264, 3211264, 401408]
total_extra_time = [-4.0069, 0.5731, 1.9653, -0.5360]

# Create a figure with one subplot
fig, ax = plt.subplots(figsize=(12, 4), dpi=300)

# Set the width of the bars
bar_width = 0.35

# Calculate the positions of the bars
x = np.arange(len(layers))
sdc_pos = x - bar_width/2
time_pos = x + bar_width/2

# Plot SDC probability for selected layers
ax.bar(sdc_pos, sdc_prob, width=bar_width, alpha=0.7, color='blue', label='SDC Probability')
ax.set_xlabel('Layer')
ax.set_ylim(-5, 2)  # Adjust the y-axis limits to accommodate the total extra time
ax.legend(loc='upper left')

# Plot total extra time for selected layers
ax.bar(time_pos, total_extra_time, width=bar_width, alpha=0.7, color='red', label='Total Extra Time')
ax.legend(loc='upper right')

# Set the x-tick labels to the selected layer names
ax.set_xticks(x)
ax.set_xticklabels(layers)

# Add data size as text labels above the y-axis line at y=2
for i, v in enumerate(data_size):
    ax.text(i, 2.05, f'Data Size: {v}', ha='center', fontsize=8, va='bottom')

# Add layer type as text labels below the bars
for i, v in enumerate(layer_types):
    ax.text(i, -4.8, v, ha='center', fontsize=8)

# Draw a horizontal line at y=0
ax.axhline(y=0, color='black', linewidth=0.8)

# Adjust layout
plt.tight_layout()

# Save the plot as a high DPI image
plt.savefig('representative_layers.png', dpi=300)

# Display the plot
plt.show()
```
